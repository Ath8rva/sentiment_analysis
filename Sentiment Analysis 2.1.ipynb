```python
#Get Channel Details
request = youtube.channels().list(part = 'statistics', forUsername = 'TheEllenShow')

response = request.execute()

for i in response:
    if i == 'items':
        for j in response[i][0]:
            print(j, '=', response[i][0][j])
    else:
        print(i, '=', response[i])
```
